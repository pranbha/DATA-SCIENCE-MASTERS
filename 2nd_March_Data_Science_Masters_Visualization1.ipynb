```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y, c='blue', alpha = .99, marker='.')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("X vs Y")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y, c='blue', alpha = .99, marker='>')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("X vs Y")
plt.grid()
```

```python
plt.figure(figsize=(8,6))
x=np.linspace(1,10, 100)
y=np.sin(x)
plt.plot(x,y, '--', c='red')
plt.xlabel("Normal Values")
plt.ylabel("Sin value of X")
plt.title("Relation of X and Y")
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
y
```

```python
plt.figure(figsize = (6,6))
plt.bar(x,y, color = 'b')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid()
```

```python
plt.figure(figsize = (6,6))
plt.barh(x,y, color = 'r')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid()
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.figure(figsize=(3,3))
plt.plot(x,y)
```

```python
plt.figure(figsize=(6,4))
data=[1,2,3,4,5,6,7,8,4,3,2,5,6,7,8,8,8]
plt.hist(data, color='red')
plt.xlabel("Representing unique values")
plt.ylabel("Frequency of X values")
plt.title("Histogram")
plt.grid()
plt.savefig("hist_example.png")
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z, c='green', alpha=0.9)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
#plt.zlabel("Z-Axis")
```
